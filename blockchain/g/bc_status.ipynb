### 共识算法比较

- PoW
  
  找到一个nonce使得区块B和哈希函数H满足公式
  $$H(B, nonce) ≤ m$$
  
- PoS

  通过引入账户余额以及币龄简化了区块难度
  
  $$H(H(B_{prev},A,t)) ≤ balance(A)m$$
  
- DPoS

  引入代理机制，选择一小群代理进行区块创建，验证及监督，大幅减少了时间算力成本，确定的维护链

综合一下：

| | 产生区块难度 | 分叉 | 贿赂攻击 | 长距离攻击 | 币龄加和攻击| 预计算攻击 | DDos | 女巫攻击 |
|---|---|---|---|---|---|---|---|---|
| PoW | 难   | 少 | 难 |- |-|-|-|有|
| PoS | 容易 | 多 | 有 | 有 |有|有 | 有 |有|
| DPoS| 容易 | 少 |-| 有 | 有 |-|-| 有|

### 区块链系统目标

 - 高并发，低延迟确认
 - 兼容多个共识(跨链)
 - 共识机制，避免攻击, 可自我修复
 - 平滑升级，避免分叉
 - 可扩展的API

## 经典案例分享

 - 以太猫
 
  虚拟世界真实化。喂养，拍卖游戏宠物。
  
 - 加密名人
 
  玩法无厘头，所有人都可以通过竞价购买方式获取该名人

 - EOS
 
 以太坊加强版，使用DPOS共识算法
 
   - 基于注册账户的智能合约
   - 账户组合控制
   - 延迟消息响应，邮件&短信通知授权
   - 令牌模型
   - 更新合约
   - 跨链交互
 
 - Steem
 
  打造社交媒体项目闭环
  
  货币分类: 
   - STEEM(交易所代币)
   - SP(拥有投票权,享受增发STEEM收益)，可转STEEM，有锁定期
   - SBD(类似USDT泰达币)
   
  角色分类：
   - 平台(运营)： 发行STEEM，以及提供工具（例如推荐位）来让用户购买并消耗STEEM
   - 核心用户： 持有SP的人，享受增发90%奖励
   - 创作者： 根据支持度收取增发奖励
   - 投票者(粉丝)： 增发收益15%
   - 矿工: 增发收益10%
